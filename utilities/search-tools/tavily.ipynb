### 2) Scrap Web Page

```python
def scrap_webpage(pageurl):
    tavily_client = TavilyClient()

    urls = [ pageurl ]

    response = tavily_client.extract(urls=urls, include_images=False)
    return response["results"][0]['raw_content']
```

```python
page = "https://idrw.org/"
print(scrap_webpage(page))
```
